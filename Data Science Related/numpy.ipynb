```python
import random
n = []
for i in range(1000000):
    n.append(random.randint(1,100))
```

```python
c = {}
for i in n:
    if(i in c):
        c[i]+=1
    else:
        c[i]=1
c
```

```python
import matplotlib.pyplot as plt
plt.plot(c.values());
plt.ylim(0,20000)
```

```python
Gr = i.max(axis=2)
plt.imshow(Gr, "gray")
plt.show()
```

```python
Gr.shape            #بعد سوم 1 شد مثل اینکه یک لایه از یک مکعب برش بدیم پس بعد سوم  یک لایه میشه
```

```python
Gr = i.sum(axis=2)           #grayscale with sum
plt.imshow(Gr, "gray")
plt.show()
```

```python
Gr = i[:, :, 0] + i[:, :, 1] + i[:, :, 2]    # overflow happens !!!!!!!!!!!!!! more than 255
plt.imshow(Gr, "gray")
plt.show()
```

```python
Gr = i[:, :, 0]//3 + i[:, :, 1]//3 + i[:, :, 2] //3    # overflow not happens !!!!!!!!!!!!!!
plt.imshow(Gr, "gray")
plt.show()
```

```python
Gr = (i[:, :, 0].astype('uint64') + i[:, :, 1].astype('uint64') + i[:, :, 2].astype('uint64'))/3    
plt.imshow(Gr, "gray")           #overflow not happen because of changing type
plt.show()
```

```python
Gr = i.mean(axis=2)              # easiest way
plt.imshow(Gr, "gray")
plt.show()
```

```python
# correct and most accurate way 
m = np.array([0.2126, 0.7152, 0.0722])
Gr = i[:, :, 0] * 0.2126 + i[:, :, 1] * 0.7152 + i[:, :, 2] * 0.0722
plt.imshow(Gr, "gray")
plt.show()
```
