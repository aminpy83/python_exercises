```python
x = np.linspace(0,360,50)
y = np.sin(np.radians(x))
y2 = np.cos(np.radians(x))
```

```python
plt.plot(x,y);    #plot for data series ترتیبی
plt.show()
```

```python
plt.plot(y) # generated x by arange()
plt.show()
```

```python
plt.plot(x,y, '--')
plt.plot(x,y2, 'ro')
plt.show()
```

```python
plt.plot(x,y, '--')
plt.plot(x,y2, 'ro')
plt.legend(['$sin(x)$', '$cos(x)$'])
plt.title('title here')
plt.xlabel('values')
plt.ylabel(' y values')
plt.show()
```

```python
plt.plot(x,y, '--')
plt.plot(x,y2, 'ro')
plt.grid(which='both')
plt.show()
```

```python
x = np.linspace(0,360,50)
y = np.sin(np.radians(x))
y2 = np.cos(np.radians(x))

plt.figure(figsize=(15, 8))

plt.plot(x, y, 'g')
plt.plot(x, y2, 'r');
```

```python
x = np.linspace(0,360,50)
y = np.sin(np.radians(x))
y2 = np.cos(np.radians(x))

plt.figure(figsize=(15, 8), dpi=20) # dpi : در هر اینچ چند نقطه بکشه

plt.plot(x, y, 'g')
plt.plot(x, y2, 'r');
```

```python
x = np.linspace(0,360,50)
y = np.sin(np.radians(x))
y2 = np.cos(np.radians(x))

plt.figure(figsize=(15, 8), dpi=300) # dpi : در هر اینچ چند نقطه بکشه

plt.plot(x, y, 'g')
plt.plot(x, y2, 'r');
```

```python
x = np.linspace(0,360,50)
y = np.sin(np.radians(x))
y2 = np.cos(np.radians(x))
plt.figure(figsize=(15, 8), dpi=72) # dpi : در هر اینچ چند نقطه بکشه

plt.subplot(1, 2, 1)      # زیر نمودار ( سطر و ستون و کدوم)
plt.plot(x, y, 'g')
plt.subplot(1, 2, 2)
plt.plot(x, y2, 'r');
```
